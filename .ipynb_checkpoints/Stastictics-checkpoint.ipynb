```python
# 2. Generate a list o> 500 integers containing values between 200 to 300 and store it in the variable `int_list2`.


import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# 2. Visualizations

# Frequency & Gaussian distribution
plt.figure(figsize=(10, 6))
sns.histplot(int_list2, kde=False, bins=30, color='blue', alpha=0.7, label='Frequency')
mu, std = np.mean(int_list2), np.std(int_list2)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, std)
plt.plot(x, p * (500/30), 'k', linewidth=2, label='Gaussian Distribution')
plt.title('Frequency & Gaussian Distribution')
plt.legend()
plt.show()

# Frequency smoothened KDE plot
plt.figure(figsize=(10, 6))
sns.histplot(int_list2, kde=True, bins=30, color='green', alpha=0.7, label='Frequency')
plt.title('Frequency Smoothened KDE Plot')
plt.legend()
plt.show()

# Gaussian distribution & smoothened KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(int_list2, color='red', label='KDE')
plt.plot(x, p, 'k', linewidth=2, label='Gaussian Distribution')
plt.title('Gaussian Distribution & Smoothened KDE Plot')
plt.legend()
plt.show()

# 3. Statistical Functions

# (ii) Function to calculate the range of a given list of numbers
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# (iii) Function to find the variance and standard deviation of a list of numbers
def calculate_variance_std(numbers):
    variance = np.var(numbers)
    std_dev = np.std(numbers)
    return variance, std_dev

# (iv) Function to compute the interquartile range (IQR) of a list of values
def calculate_iqr(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return q3 - q1

# (v) Function to calculate the coefficient of variation for a dataset
def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers)
    return std_dev / mean

# (vi) Function to find the mean absolute deviation (MAD) of a list of numbers
def calculate_mad(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(numbers - mean))

# (vii) Function to calculate the quartile deviation of a list of values
def calculate_quartile_deviation(numbers):
    q1 = np.percentile(numbers, 25)
    q3 = np.percentile(numbers, 75)
    return (q3 - q1) / 2

# (viii) Function to find the range-based coefficient of dispersion for a dataset
def calculate_range_coefficient_of_dispersion(numbers):
    range_value = calculate_range(numbers)
    mean = np.mean(numbers)
    return range_value / mean

# Sample calculations
range_value = calculate_range(int_list2)
variance, std_dev = calculate_variance_std(int_list2)
iqr = calculate_iqr(int_list2)
coefficient_of_variation = calculate_coefficient_of_variation(int_list2)
mad = calculate_mad(int_list2)
quartile_deviation = calculate_quartile_deviation(int_list2)
range_coefficient_of_dispersion = calculate_range_coefficient_of_dispersion(int_list2)

print("Range:", range_value)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
print("Interquartile Range (IQR):", iqr)
print("Coefficient of Variation:", coefficient_of_variation)
print("Mean Absolute Deviation (MAD):", mad)
print("Quartile Deviation:", quartile_deviation)
print("Range-based Coefficient of Dispersion:", range_coefficient_of_dispersion)
```

```python
# 3)  Write a Python class representing a discrete random variable with methods to calculate its expected
# value and variance.

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initialize the discrete random variable with values and their corresponding probabilities.
        :param values: List of possible values of the random variable.
        :param probabilities: List of corresponding probabilities for each value.
        """
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not all(0 <= p <= 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if not abs(sum(probabilities) - 1.0) < 1e-6:
            raise ValueError("Probabilities must sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculate the expected value (mean) of the random variable.
        :return: Expected value.
        """
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculate the variance of the random variable.
        :return: Variance.
        """
        mean = self.expected_value()
        return sum((value - mean) ** 2 * prob for value, prob in zip(self.values, self.probabilities))

# Example usage
values = [0, 1, 2, 3, 4]  # Possible values of the random variable
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]  # Corresponding probabilities

rv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
```

```python
    # _) Create a Python function to generate random samples 6rom a given probabilits distribution (e.g.,
    # binomial, Poisson) and calculate their mean and variance.

    import numpy as np

    def generate_samples_and_calculate_stats(distribution, params, sample_size=1000):
        """
        Generate random samples from a given distribution and calculate their mean and variance.
        :param distribution: The name of the distribution ('binomial' or 'poisson').
        :param params: A dictionary of parameters required for the distribution.
        :param sample_size: The number of samples to generate (default is 1000).
        :return: A tuple containing the mean and variance of the samples.
        """
        if distribution == 'binomial':
            n = params.get('n')
            p = params.get('p')
            if n is None or p is None:
                raise ValueError("For binomial distribution, 'n' and 'p' parameters are required.")
            samples = np.random.binomial(n, p, sample_size)
        elif distribution == 'poisson':
            lam = params.get('lam')
            if lam is None:
                raise ValueError("For poisson distribution, 'lam' parameter is required.")
            samples = np.random.poisson(lam, sample_size)
        else:
            raise ValueError("Unsupported distribution. Supported distributions are 'binomial' and 'poisson'.")

        mean = np.mean(samples)
        variance = np.var(samples)
        return mean, variance

    # Example usage
    # Binomial distribution with n=10 and p=0.5
    binomial_params = {'n': 10, 'p': 0.5}
    mean_binomial, variance_binomial = generate_samples_and_calculate_stats('binomial', binomial_params)
    print("Binomial Distribution - Mean:", mean_binomial, "Variance:", variance_binomial)

    # Poisson distribution with =3
    poisson_params = {'lam': 3}
    mean_poisson, variance_poisson = generate_samples_and_calculate_stats('poisson', poisson_params)
    print("Poisson Distribution - Mean:", mean_poisson, "Variance:", variance_poisson)
```

```python

# Question 7: Use the Seaborn library to load the tips dataset. Find the following from the dataset for the columns 'total_bill' and 'tip':
# (a) Write a Python function that calculates their skewness.
# (b) Create a program that determines whether the columns exhibit positive skewness, negative skewness, or are approximately symmetric.
# (c) Write a function that calculates the covariance between two columns.
# (d) Implement a Python program that calculates the Pearson correlation coefficient between two columns.
# (e) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using scatter plots.import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew


# Load the tips dataset using the Seaborn library
df = sns.load_dataset('tips')

# Function to calculate skewness
def calculate_skewness(column):
    return skew(column)

# Calculate skewness for 'total_bill' and 'tip'
total_bill_skewness = calculate_skewness(df['total_bill'])
tip_skewness = calculate_skewness(df['tip'])

print(f"Skewness of 'total_bill': {total_bill_skewness}")
print(f"Skewness of 'tip': {tip_skewness}")

# Function to determine skewness type
def determine_skewness_type(skewness):
    if skewness > 0:
        return "Positive skewness"
    elif skewness < 0:
        return "Negative skewness"
    else:
        return "Approximately symmetric"

print(f"'total_bill' skewness type: {determine_skewness_type(total_bill_skewness)}")
print(f"'tip' skewness type: {determine_skewness_type(tip_skewness)}")

# Function to calculate covariance between two columns
def calculate_covariance(col1, col2):
    return np.cov(col1, col2)[0, 1]

# Calculate covariance between 'total_bill' and 'tip'
covariance = calculate_covariance(df['total_bill'], df['tip'])
print(f"Covariance between 'total_bill' and 'tip': {covariance}")

# Function to calculate Pearson correlation coefficient
def calculate_pearson_correlation(col1, col2):
    return np.corrcoef(col1, col2)[0, 1]

# Calculate Pearson correlation coefficient between 'total_bill' and 'tip'
pearson_corr = calculate_pearson_correlation(df['total_bill'], df['tip'])
print(f"Pearson correlation coefficient between 'total_bill' and 'tip': {pearson_corr}")

# Visualize the correlation using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['total_bill'], df['tip'], alpha=0.6)
plt.title('Scatter plot of total_bill vs. tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

```python
# Question 8: Write a Python function to calculate the Probability Density Function (PDF) of a continuous random variable
# for a given normal distribution.

import scipy.stats as stats


def calculate_pdf(value, mean, std_dev):
    """
    Calculate the Probability Density Function (PDF) of a continuous random variable
    for a given normal distribution.

    Parameters:
    value (float): The point at which to evaluate the PDF.
    mean (float): The mean of the normal distribution.
    std_dev (float): The standard deviation of the normal distribution.

    Returns:
    float: The PDF value at the given point.
    """
    # Calculate the PDF using scipy.stats.norm
    pdf_value = stats.norm.pdf(value, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean = 0       # Mean of the normal distribution
std_dev = 1    # Standard deviation of the normal distribution
value = 1.5    # Point at which to evaluate the PDF

pdf_value = calculate_pdf(value, mean, std_dev)
print(f"PDF at x = {value}: {pdf_value}")
```

```python
# Question 9: Create a program to calculate the Cumulative Distribution Function (CDF) of an exponential distribution.
import scipy.stats as stats



def calculate_cdf(value, rate):
    """
    Calculate the Cumulative Distribution Function (CDF) of an exponential distribution.

    Parameters:
    value (float): The point at which to evaluate the CDF.
    rate (float): The rate () of the exponential distribution, which is the reciprocal of the mean.

    Returns:
    float: The CDF value at the given point.
    """
    # Calculate the CDF using scipy.stats.expon
    cdf_value = stats.expon.cdf(value, scale=1/rate)
    return cdf_value

# Example usage
rate = 1.0   # Rate () of the exponential distribution
value = 2.0   # Point at which to evaluate the CDF

cdf_value = calculate_cdf(value, rate)
print(f"CDF at x = {value}: {cdf_value}")
```

```python
# Question 10: Write a Python function to calculate the Probability Mass Function (PMF) of a Poisson distribution.

import scipy.stats as stats



def calculate_pmf(k, rate):
    """
    Calculate the Probability Mass Function (PMF) of a Poisson distribution.

    Parameters:
    k (int): The point at which to evaluate the PMF (k must be a non-negative integer).
    rate (float): The rate () of the Poisson distribution, which is the average number of occurrences in a fixed interval.

    Returns:
    float: The PMF value at the given point.
    """
    # Calculate the PMF using scipy.stats.poisson
    pmf_value = stats.poisson.pmf(k, mu=rate)
    return pmf_value

# Example usage
rate = 3.0  # Rate () of the Poisson distribution
k = 4       # Point at which to evaluate the PMF

pmf_value = calculate_pmf(k, rate)
print(f"PMF at k = {k}: {pmf_value}")
```
